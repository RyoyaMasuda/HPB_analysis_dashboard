```python
#   カラー率を計算したい
```

```python
dropdown1_value='島根'
dropdown2_value='松江'
dropdown3_value='SMITH AVEDA'
_df2 = _df.copy()
_df2 = _df2[_df2['県']==dropdown1_value]
_df2 = _df2[_df2['エリア'] == dropdown2_value]
_df2 = _df2[_df2['サロン名'] == dropdown3_value]
_df2
```

```python
_df2['コメントへの返信'].value_counts(normalize=True)
```

```python
round((1 - _df2['コメントへの返信'].value_counts(normalize=True)['無']), 3)
```

```python
for prefecture in tqdm(_df['県'].unique()):
    tmp_df = _df.copy()
    tmp_df = tmp_df[tmp_df['県']==prefecture]
    for area in tmp_df['エリア'].unique():
        tmp_df2 = tmp_df.copy()
        tmp_df2 = tmp_df2[tmp_df2['エリア']==area]
        for salon_name in tmp_df2['サロン名'].unique():
            tmp_df3 = tmp_df2.copy()
            tmp_df3 = tmp_df3[tmp_df3['サロン名']==salon_name]
            try:
                reply_rate = round((tmp_df3['コメントへの返信'].value_counts(normalize=True)['有']), 3)
            except KeyError:
                reply_rate = 0.0
            _df.loc[((_df['県']==prefecture)&(_df['エリア']==area)&(_df['サロン名']==salon_name)), 'コメントへの返信率']=reply_rate
```
